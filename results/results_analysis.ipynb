```python
import os
import pickle
import pandas as pd


if not os.path.exists("../formatted_results/COMA_ground_associated"):
    os.mkdir("../formatted_results/COMA_ground_associated")
if not os.path.exists("../formatted_results/Word2Vec_ground_associated"):
    os.mkdir("../formatted_results/Word2Vec_ground_associated")
```

```python


for res_file in [f for f in os.listdir(".") if
                      os.path.isfile(os.path.join(".", f)) and f.endswith("dict.txt")]:
    res_dict_whole = pickle.load(open("../results/"+res_file,"br"))
    # print(res_file)
    if res_file.split("_")[0] == "Word2Vec":
        if res_file.split("_")[1] == "FHIR" and res_file.split("_")[3]=="C-CDA":

            ground_truth=pd.read_csv("../mappings/FHIR_to_C-CDA_._delimiter.csv")
            for expe in res_dict_whole:

                res_dict = res_dict_whole[expe]
                list_res=[("key","match","dist")]
                for key in res_dict["results"].keys():
                    list_res.append((key, res_dict["results"][key][0], res_dict["results"][key][1]))
                pd.DataFrame(list_res).to_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv", index=False, header=False)

                matching = pd.read_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv")
                res={"not_in_groundtruth":[], "in_groundtruth":{}}
                for match,key,dist in zip(matching["match"],matching["key"],matching["dist"]):
                    key_present = ground_truth[ground_truth["key"].map(lambda x : x.endswith(key))]
                    # print(key_present)
                    # print(key_present)
                    # if key =="system":
                    #     print(set(list(key_present["key"])),(list(key_present["key"])), key, matching)
                    # if len(set(list(key_present["key"]))) == 1:
                    res["in_groundtruth"][(key,match,dist)]= [list(key_present["match"])]
                    # else:
                    #     res["not_in_groundtruth"].append((key,match,dist))


                pd.DataFrame(res["in_groundtruth"]).to_csv("../formatted_results/aux.csv")
                aux = pd.read_csv("../formatted_results/aux.csv")
                aux["Unnamed: 0"]=["match_done","dist","match_that_should_be_done"]
                aux = aux.rename(columns={"Unnamed: 0":"key"})
                aux = aux.transpose()
                aux.to_csv("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv", header=None)
        if res_file.split("_")[1] == "ODATIS" and res_file.split("_")[3]=="AERIS":
            for expe in res_dict_whole:
                # ground_truth=ground_truth_ODATIS_AERIS
                ground_truth=pd.read_csv("../mappings/ODATIS_to_AERIS_._delimiter_only_leaf.csv.csv")
                res_dict = res_dict_whole[expe]
                list_res=[("key","match","dist")]
                for key in res_dict["results"].keys():
                    list_res.append((key, res_dict["results"][key][0], res_dict["results"][key][1]))
                pd.DataFrame(list_res).to_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv", index=False, header=False)

                matching = pd.read_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv")
                res={"not_in_groundtruth":[], "in_groundtruth":{}}
                for match,key,dist in zip(matching["match"],matching["key"],matching["dist"]):
                    key_present = ground_truth[ground_truth["key"].map(lambda x : x.endswith(key))]


                    # print(key_present)
                    # if key =="system":
                    #     print(set(list(key_present["key"])),(list(key_present["key"])), key, matching)
                    # if len(set(list(key_present["key"]))) == 1:
                    res["in_groundtruth"][(key,match,dist)]= [list(key_present["match"])]
                    # else:
                        # res["not_in_groundtruth"].append((key,match,dist))


                pd.DataFrame(res["in_groundtruth"]).to_csv("../formatted_results/aux.csv")
                aux = pd.read_csv("../formatted_results/aux.csv")
                aux["Unnamed: 0"]=["match_done","dist","match_that_should_be_done"]
                aux = aux.rename(columns={"Unnamed: 0":"key"})
                aux = aux.transpose()
                # print("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv")
                aux.to_csv("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv", header=None)
        if res_file.split("_")[1] == "AERIS" and res_file.split("_")[3]=="FHIR":
            for expe in res_dict_whole:
                # ground_truth=ground_truth_ODATIS_AERIS
                ground_truth=pd.read_csv("../mappings/AERIS_to_FHIR_._delimiter.csv")
                res_dict = res_dict_whole[expe]
                list_res=[("key","match","dist")]
                for key in res_dict["results"].keys():
                    list_res.append((key, res_dict["results"][key][0], res_dict["results"][key][1]))
                pd.DataFrame(list_res).to_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv", index=False, header=False)

                matching = pd.read_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv")
                res={"not_in_groundtruth":[], "in_groundtruth":{}}
                for match,key,dist in zip(matching["match"],matching["key"],matching["dist"]):
                    key_present = ground_truth[ground_truth["key"].map(lambda x : x.endswith(key))]


                    # print(key_present)
                    # if key =="system":
                    #     print(set(list(key_present["key"])),(list(key_present["key"])), key, matching)
                    # if len(set(list(key_present["key"]))) == 1:
                    res["in_groundtruth"][(key,match,dist)]= [list(key_present["match"])]
                    # else:
                        # res["not_in_groundtruth"].append((key,match,dist))


                pd.DataFrame(res["in_groundtruth"]).to_csv("../formatted_results/aux.csv")
                aux = pd.read_csv("../formatted_results/aux.csv")
                aux["Unnamed: 0"]=["match_done","dist","match_that_should_be_done"]
                aux = aux.rename(columns={"Unnamed: 0":"key"})
                aux = aux.transpose()
                # print("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv")
                aux.to_csv("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv", header=None)


    if res_file.split("_")[0] == "COMA":
        if res_file.split("_")[1] == "FHIR" and res_file.split("_")[3]=="C-CDA":
            ground_truth=pd.read_csv("../mappings/FHIR_to_C-CDA_._delimiter.csv")
            for expe in res_dict_whole:

                res_dict = res_dict_whole[expe]
                list_res=[("key","match","dist")]
                # print(res_dict["results"])

                for key in res_dict["results"].keys():

                    list_res.append((key[0][1],key[1][1], res_dict["results"][key]))
                pd.DataFrame(list_res).to_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv", index=False, header=False)

                matching = pd.read_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv")
                res={"not_in_groundtruth":[], "in_groundtruth":{}}
                for match,key,dist in zip(matching["match"],matching["key"],matching["dist"]):
                    key_present = ground_truth[ground_truth["key"].map(lambda x : x.endswith(key))]
                    # print(key_present)
                    # if key =="system":
                    #     print(set(list(key_present["key"])),(list(key_present["key"])), key, matching)
                    # if len(set(list(key_present["key"]))) == 1:
                    res["in_groundtruth"][(key,match,dist)]= [list(key_present["match"])]
                    # else:
                    #     res["not_in_groundtruth"].append((key,match,dist))


                pd.DataFrame(res["in_groundtruth"]).to_csv("../formatted_results/aux.csv")
                aux = pd.read_csv("../formatted_results/aux.csv")
                aux["Unnamed: 0"]=["match_done","dist","match_that_should_be_done"]
                aux = aux.rename(columns={"Unnamed: 0":"key"})
                aux = aux.transpose()
                aux.to_csv("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv", header=None)

        if res_file.split("_")[1] == "ODATIS" and res_file.split("_")[3]=="AERIS":
            ground_truth=pd.read_csv("../mappings/ODATIS_to_AERIS_._delimiter_only_leaf.csv.csv")
            for expe in res_dict_whole:


                res_dict = res_dict_whole[expe]
                list_res=[("key","match","dist")]
                # print(res_dict["results"])

                for key in res_dict["results"].keys():
                    # print(key)
                    list_res.append((key[0][1],key[1][1], res_dict["results"][key]))
                pd.DataFrame(list_res).to_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv", index=False, header=False)

                matching = pd.read_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv")
                res={"not_in_groundtruth":[], "in_groundtruth":{}}
                for match,key,dist in zip(matching["match"],matching["key"],matching["dist"]):
                    key_present = ground_truth[ground_truth["key"].map(lambda x : x.endswith(key))]
                    # print(key_present)
                    # if key =="system":
                    #     print(set(list(key_present["key"])),(list(key_present["key"])), key, matching)
                    # if len(set(list(key_present["key"]))) == 1:
                    res["in_groundtruth"][(key,match,dist)]= [list(key_present["match"])]
                    # else:t(res_dict
                    #     res["not_in_groundtruth"].append((key,match,dist))


                pd.DataFrame(res["in_groundtruth"]).to_csv("../formatted_results/aux.csv")
                aux = pd.read_csv("../formatted_results/aux.csv")
                aux["Unnamed: 0"]=["match_done","dist","match_that_should_be_done"]
                aux = aux.rename(columns={"Unnamed: 0":"key"})
                aux = aux.transpose()
                aux.to_csv("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv", header=None)
        if res_file.split("_")[1] == "AERIS" and res_file.split("_")[3]=="FHIR":
            ground_truth=pd.read_csv("../mappings/AERIS_to_FHIR_._delimiter.csv")
            for expe in res_dict_whole:

                res_dict = res_dict_whole[expe]
                list_res=[("key","match","dist")]


                for key in res_dict["results"].keys():

                    list_res.append((key[0][1],key[1][1], res_dict["results"][key]))

                pd.DataFrame(list_res).to_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv", index=False, header=False)

                matching = pd.read_csv("../formatted_results/unformatted_results/"+expe.replace("/","_")+res_file+".csv")
                res={"not_in_groundtruth":[], "in_groundtruth":{}}
                for match,key,dist in zip(matching["match"],matching["key"],matching["dist"]):
                    key_present = ground_truth[ground_truth["key"].map(lambda x : x.endswith(key))]
                    # print(key_present)
                    # if key =="system":
                    #     print(set(list(key_present["key"])),(list(key_present["key"])), key, matching)
                    # if len(set(list(key_present["key"]))) == 1:
                    res["in_groundtruth"][(key,match,dist)]= [list(key_present["match"])]
                    # else:
                    #     res["not_in_groundtruth"].append((key,match,dist))


                pd.DataFrame(res["in_groundtruth"]).to_csv("../formatted_results/aux.csv")
                aux = pd.read_csv("../formatted_results/aux.csv")
                aux["Unnamed: 0"]=["match_done","dist","match_that_should_be_done"]
                aux = aux.rename(columns={"Unnamed: 0":"key"})
                aux = aux.transpose()
                aux.to_csv("../formatted_results/"+res_file.split("_")[0]+"_ground_associated/"+expe.replace("/","_")+"_with_matching_"+res_file+".csv", header=None)
```
